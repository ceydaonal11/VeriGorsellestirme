# Heatmap

**Veri Seti Hikayesi**

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.shape
```

```python
df["passengers"].describe()
```

```python
df = df.pivot("month", "year","passengers");
```

```python
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot = True, fmt="d");
```

```python
sns.heatmap(df, annot = True, fmt="d", linewidths= .5);
```
